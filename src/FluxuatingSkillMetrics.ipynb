```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_json("deep_QL_Train_winners.json")
```

```python
total_winner = df.to_numpy()
```

```python
total_winner[:100].sum()
```

```python
total_winner[900:].sum()
```

```python
import matplotlib.pyplot as plt

# assume 'rewards' is your list of numbers (1, 0, -1)

cumulative_reward = 0
cumulative_rewards = [total_winner[0]]
for i in range(1, len(total_winner)):
    cumulative_rewards.append(cumulative_rewards[-1] + total_winner[i])
```

```python
plt.plot(cumulative_rewards)
plt.xlabel('Time Step')
plt.ylabel('Cumulative Reward')
plt.title('Cumulative Reward Over Time')
plt.show()

# or

```

```python
win_count = 0
win_rates = []
for i, reward in enumerate(total_winner):
    if reward == 1:
        win_count += 1
    win_rate = win_count / (i + 1)
    win_rates.append(win_rate)

```

```python
plt.plot(win_rates)
plt.xlabel('Time Step')
plt.ylabel('Win Rate')
plt.title('Win Rate Over Time')
plt.show()

# or

```

```python
window_size = 10
moving_averages = []
for i in range(len(total_winner) - window_size + 1):
    window_rewards = total_winner[i:i + window_size]
    moving_average = sum(window_rewards) / window_size
    moving_averages.append(moving_average)
```

```python
plt.plot(moving_averages)
plt.xlabel('Time Step')
plt.ylabel('Moving Average Reward')
plt.title('Moving Average Reward Over Time')
plt.show()
```
